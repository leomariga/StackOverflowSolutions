```python
import os
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt


def find_coefs(original_coords, warped_coords):
        matrix = []
        for p1, p2 in zip(original_coords, warped_coords):
            matrix.append([p1[0], p1[1], 1, 0, 0, 0, -p2[0]*p1[0], -p2[0]*p1[1]])
            matrix.append([0, 0, 0, p1[0], p1[1], 1, -p2[1]*p1[0], -p2[1]*p1[1]])

        A = np.matrix(matrix, dtype=np.float)
        B = np.array(warped_coords).reshape(8)

        res = np.dot(np.linalg.inv(A.T * A) * A.T, B)
        return np.array(res).reshape(8)


def warp_keypoints(keypoints, coeffs):
        a, b, c, d, e, f, g, h = coeffs
        new_keypoints = keypoints.copy()
        for idx, keypoint in enumerate(keypoints):
            x,y = keypoint
            x = int(x)
            y = int(y)
            
            new_x = (a * x + b * y + c) / (g * x + h * y + 1)
            new_y = (d * x + e * y + f) / (g * x + h * y + 1)

            new_keypoints[idx] = (int(new_x),int(new_y)) # after transformation

        return new_keypoints


coefs = find_coefs(
                  [(867,652), (1020,580), (1206,666), (1057,757)],
                  [(700,732), (869,754), (906,916), (712,906)]
                  )

coefs_inv = find_coefs(
                  [(700,732), (869,754), (906,916), (712,906)],
                  [(867,652), (1020,580), (1206,666), (1057,757)]
                  )
print(coefs)
print(coefs_inv)

# The lines below are the Openc-CV equivalent transformation
# t_matrix = np.array([[coefs[0],coefs[1],coefs[2]],[coefs[3],coefs[4],coefs[5]],[coefs[6], coefs[7], 1]])
# print(t_matrix)
# inverse = np.linalg.inv(t_matrix)
# print(inverse/inverse[2][2])

image = Image.open('sample.png')

img = image.transform(((1500,800)),
                      method=Image.PERSPECTIVE,
                      data=coefs_inv)

# note, y coordinate is first, then x
keypoints = np.array([[136.,  72.],
                      [250,  420],
                      [350, 200],
                      [750., 370],
                      [50., 100.],
                      [400., 500.]])

new_points = warp_keypoints(keypoints, coefs)

plt.figure()
plt.imshow(image)
plt.scatter(keypoints[:, 0], keypoints[:, 1], s=100, marker='.', c='b')
plt.show()


plt.figure()
plt.imshow(img)
plt.scatter(new_points[:, 0], new_points[:, 1], s=100, marker='.', c='r')
plt.show()
```
